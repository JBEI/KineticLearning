# Create CSVs with Properly Formatted Data for Use in Kinetic Learning

Specifically Create a CSV for Isopentenol Strains and Isopentenol Strains

```python
import pandas as pd
from IPython.display import display
```

```python
#Create Data Set CSVs in Correct Format From Provided Data
#simulated_df = pd.read_csv('data/d')
metabolite_df = pd.read_csv('data/time_series_metabolomics.csv')
protein_df = pd.read_csv('data/time_series_proteomics.csv')

metadata = ['Hour','Strain']
metabolites = ['Acetyl-CoA (uM)','HMG-CoA (uM)','Intracellular Mevalonate (uM)','Mev-P (uM)','IPP/DMAPP (uM)','Limonene g/L',
               'OD600','GPP (uM)','NAD (uM)','NADP (uM)','Acetate g/L','Pyruvate g/L','citrate (uM)','Isopentenol g/L']
metabolite_names = ['Acetyl-CoA','HMG-CoA','Mevalonate','Mev-P','IPP/DMAPP','Limonene',
               'OD600','GPP','NAD','NADP','Acetate','Pyruvate','citrate','Isopentenol']
proteins = ['AtoB','Limonene Synthase','HMGR','HMGS','MK','PMK','PMD','NudB','Idi','GPPS']

#display(metabolite_df[metadata+metabolites])
#print(protein_df['Protein'].unique())
protein_df = protein_df[['Strain','Hour','Protein','ProteinArea']].loc[protein_df['Protein'].isin(proteins)]
protein_df = protein_df.loc[protein_df['Hour']!='72C']
protein_df['Hour'] = protein_df['Hour'].apply(float)
protein_df = pd.pivot_table(protein_df,values='ProteinArea',index=['Strain','Hour'],columns='Protein')

metabolite_df = metabolite_df[metadata+metabolites].set_index(['Strain','Hour'])
metabolite_df.columns = metabolite_names

#Merge Data Frames
df = pd.merge(protein_df, metabolite_df,left_index=True, right_index=True,how='outer')
display(protein_df.index)
display(metabolite_df.index)
```

```python
#Save out CSV for both Limonene & Isopentenol
limonene_strains = ['L1','L2','L3']
limonene_df = df.loc[df.index.get_level_values(0).isin(limonene_strains)]
limonene_df.to_csv('data/limonene_data.csv')
display(limonene_df)

isopentenol_strains = ['I1','I2','I3']
isopentenol_df = df.loc[df.index.get_level_values(0).isin(isopentenol_strains)]
display(isopentenol_df)
isopentenol_df.to_csv('data/isopentenol_data.csv')
```
